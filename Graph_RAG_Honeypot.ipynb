````python
%%writefile app.py
import streamlit as st
import subprocess
from pyngrok import ngrok

st.set_page_config(page_title="GraphRAG Terminal Query", page_icon="ğŸ’»")

st.title("ğŸ§  GraphRAG Unix Terminal Simulation")

st.markdown("Simulate Unix commands with a GraphRAG-powered LLM system.")

instruction = st.text_area("Instruction", value="You are now a Unix OS terminal. You act and respond exactly as a Unix terminal.")
input_cmd = st.text_input("Command Input", value="admin@server:/etc$ cat /etc/passwd")

if st.button("Run Query"):
    s = f" your Instruction is {instruction} your Input:'{input_cmd}' what will be the output?"

    st.markdown(f"### ğŸ” Query Sent:\n```\n{s}\n```")

    try:
        # Run the CLI command
        result = subprocess.run(
            ["python", "-m", "graphrag", "query", "--root", "./", "--method", "global", "--query", s],
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True,
        )
        output = result.stdout if result.stdout else result.stderr
        st.markdown("### ğŸ“¥ Output:")
        st.code(output)

    except Exception as e:
        st.error(f"Something went wrong: {e}")
````

````python
import json
import networkx as nx
from sentence_transformers import SentenceTransformer, util
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch


with open("/content/unix-commands-dataset.json", "r") as f:
    data = json.load(f)


commands = [entry["input"].strip().split()[0] for entry in data if entry.get("input")]
commands = list(set(commands))  # unique commands


model_name = "sentence-transformers/all-MiniLM-L6-v2"
encoder = SentenceTransformer(model_name)
command_embeddings = encoder.encode(commands, convert_to_tensor=True)


G = nx.Graph()
for idx, cmd in enumerate(commands):
    G.add_node(cmd, embedding=command_embeddings[idx])


threshold = 0.6  # tweakable
for i in range(len(commands)):
    for j in range(i + 1, len(commands)):
        score = util.cos_sim(command_embeddings[i], command_embeddings[j]).item()
        if score >= threshold:
            G.add_edge(commands[i], commands[j], weight=score)

````

````python
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf")
llama = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf", torch_dtype=torch.float16).eval()

llm_pipeline = pipeline("text-generation", model=llama, tokenizer=tokenizer, device=0)

# === Graph-RAG Retrieval Function ===
def query_graph_rag(query, top_k=5):
    query_vec = encoder.encode(query, convert_to_tensor=True)
    similarities = util.cos_sim(query_vec, command_embeddings)[0]
    top_indices = torch.topk(similarities, top_k).indices.tolist()

    retrieved_cmds = [commands[i] for i in top_indices]
    prompt_context = "\n".join(f"- {cmd}" for cmd in retrieved_cmds)

    prompt = f"""You are a helpful Unix assistant. Based on these related commands:
{prompt_context}

Answer the question: "{query}".
"""
    result = llm_pipeline(prompt, max_new_tokens=100, do_sample=False)[0]["generated_text"]
    return result

# === Example Query ===
response = query_graph_rag("How can I find my current working directory?")
print(response)
````
